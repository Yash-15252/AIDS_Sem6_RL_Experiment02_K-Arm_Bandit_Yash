# Problem Statement:
*An advertising comapny is running 10 different ads targeted towards a similar set of population on a webpage. We have results for which ads were clicked by a user here. Each column index represents a different ad. we have a 1 if the ad was clicked by a user, and 0 if it was not*

#Solution: K-armed bandit problem

*   First, we will try a random selection technique, where we randomly select any ad and show it to the user. If the user click the ad, we getr paid and if not, there is no profit.

```python
#Yash Khamkar - 221A030
#Random Selection
#Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Importing the dataset
dataset = pd.read_csv('/content/Ads_Optimisation.csv')
dataset.head()
```

```python
#Yash Khamkar - 221A030
#Implementing Random Selection
import random
N = 10000
d = 10
ads_selected = []
total_reward = 0
for n in range(0, N):
  ad = random.randrange(d) #Add Random Selection
  ads_selected.append(ad)
  reward = dataset.values[n, ad]
  total_reward = total_reward + reward
```

```python
#Yash Khamkar - 221A030
#Implementing UCB
import math
import random
import matplotlib.pyplot as plt
import numpy as np
N =  10000
d = 10
ads_selected = []
numbers_of_selections = [0] * d
sums_of_rewards = [0] * d
total_reward = 0

for n in range(0, N):
  ad = 0
  max_upper_bound = 0
  for i in range(0, d):
    if (numbers_of_selections[i] > 0):
      average_reward = sums_of_rewards[i] / numbers_of_selections[i]
      delta_i = math.sqrt(2 * math.log(n + 1) / numbers_of_selections[i])
      upper_bound = average_reward + delta_i
    else:
      upper_bound = 1e400
    if upper_bound > max_upper_bound:
      max_upper_bound = upper_bound
      ad = i
  ads_selected.append(ad)
  numbers_of_selections[ad] += 1
  reward = dataset.values[n, ad]
  sums_of_rewards[ad] += reward
  total_reward += reward
```
